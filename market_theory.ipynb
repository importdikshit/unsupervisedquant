# Simple Auctions

Auctions are a common way to sell/buy goods when:
􏰀 Seller/buyer has little knowledge of what would be the “right”
price.
􏰀 There is scarcity: fixed supply (Monet painting).
􏰀 The quality/quantity of the good to be sold/buy changes very frequently (electricity, fish, etc.).
􏰀 Transaction frequency is low.


An auction is defined by
􏰀 Bidding format rules: Price only, multi-attribute, price and quantity, or quantity only.
􏰀 Bidding process rules: Closing/timing rules, available information, rules for bid improvements/counter-bids, closing conditions.
􏰀 Price and allocation rules: Final price(s) and quantities, winning bidders.


ENGLISH OUTCRY BIDDING - Depends on how the other bidders bid. IMPLICATION No dominant strategy.
